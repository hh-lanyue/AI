### torch.nn.functional 很多层和函数在这里都会见到

torch.nn.functional中有很多功能，后续会常用的。那什么时候使用nn.Module，什么时候使用nn.functional呢？一般情况下，如果模型有可学习的参数，最好用nn.Module，其他情况nn.functional相对更简单一些

```python
import torch.nn.functional as F

loss_func = F.cross_entropy

def model(xb):
    return xb.mm(weights) + bias
```

```python
# 训练个数为64，即 batch 为64
bs = 64
# 取前64个训练样本
xb = x_train[0:bs]
yb = y_train[0:bs]
# 784 为特征个数
# weights 为随机的权重参数
# 每个图像有 784 个特征，最后需要分类（10分类问题），所以中间需要一个 [784, 10] 的二维矩阵
weights = torch.randn([784, 10], dtype = torch.float,  requires_grad = True) 
bs = 64
bias = torch.zeros(10, requires_grad=True)

print(model(xb)[0])
print(loss_func(model(xb), yb))
```

### 创建一个model来更简化代码

- 必须继承nn.Module且在其构造函数中需调用nn.Module的构造函数
- 无需写反向传播函数，nn.Module能够利用autograd自动实现反向传播
- Module中的可学习参数可以通过named_parameters()或者parameters()返回迭代器

```python
from torch import nn

class Mnist_NN(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden1 = nn.Linear(784, 128)
        self.hidden2 = nn.Linear(128, 256)
        self.out  = nn.Linear(256, 10)
        # 为了防止过拟合，随机杀掉百分比的神经元
        # 基本每个全连接层都需要加一个 dropout
        self.dropout = nn.Dropout(0.5)

    # 定义前向传播路径
    def forward(self, x):
        # relu 为激活函数，激活函数的作用是增加非线性表达
        x = F.relu(self.hidden1(x))
        # 如果神经网络的参数量很大的时候，会出现过拟合的情况，Dropout 的作用是防止过拟合，丢弃一定概率的神经元
        x = self.dropout(x)
        x = F.relu(self.hidden2(x))
        x = self.dropout(x)
        x = self.out(x)
        return x
        
```

```python
net = Mnist_NN()
print(net)
```

可以打印我们定义好名字里的权重和偏置项

```python
for name, parameter in net.named_parameters():
    # 输出的 bias 为对预测结果进行微调的参数 即 y = ax + b 中的 b
    print(name)
    print(parameter.shape)
    print("=======================================")
```

### 使用TensorDataset和DataLoader来简化

```python
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

# 自定义数据集的方法（简单创建，后续不用这种）
train_ds = TensorDataset(x_train, y_train)
train_dl = DataLoader(train_ds, batch_size=bs, shuffle=True)

valid_ds = TensorDataset(x_valid, y_valid)
valid_dl = DataLoader(valid_ds, batch_size=bs * 2)

print(valid_ds)
```

```python
def get_data(train_ds, valid_ds, bs):
    return (
        DataLoader(train_ds, batch_size=bs, shuffle=True),
        DataLoader(valid_ds, batch_size=bs * 2),
    )
```

- 一般在训练模型时加上model.train()，这样会正常使用Batch Normalization和 Dropout
- 测试的时候一般选择model.eval()，这样就不会使用Batch Normalization和 Dropout

# zip 用法

```python
a1 = [1, 3, 5]
b1 = [2, 4, 6]
# 压缩
c1 = zip(a1, b1)
print(list(c1))
# 解压缩
a2, b2 = zip(*zip(a1, b1))
print(a2)
print(b2)
```

```python
import numpy as np

# opt 是优化器，梯度下降
def fit(steps, model, loss_func, opt, train_dl, valid_dl):
    for step in range(steps):
        # 要先训练-更新权重参数、偏置参数
        model.train()
        # 从封装好的数据集中拿数据
        for xb, yb in train_dl:
            loss_batch(model, loss_func, xb, yb, opt)

        # 再验证-不更新参数
        model.eval()
        # 下面一行的代码意为 "不需要计算梯度"
        with torch.no_grad():
            losses, nums = zip(
                *[
                    loss_batch(model, loss_func, xb, yb)
                    for xb, yb in valid_dl
                ]
            )
        # 计算总的损失函数
        val_loss = np.sum(np.multiply(losses, nums)) / np.sum(nums)
        print('当前step:'+str(step), '验证集损失：'+str(val_loss))
```

```python
from torch import optim
def get_model():
    model = Mnist_NN()
    # return model, optim.SGD(model.parameters(), lr=0.001)
    return model, optim.Adam(model.parameters(), lr=0.001)
```

```python
def loss_batch(model, loss_func, xb, yb, opt=None):
    # 计算损失函数
    # 更新权重参数和偏移参数
    loss = loss_func(model(xb), yb)

    if opt is not None:
        # 反向传播，计算权重参数、梯度值
        loss.backward()
        # 根据权重参数和梯度值进行更新
        opt.step()
        # torch 专属特点
        # 将上次迭代的梯度置零（因为每次迭代之间没有任何关系）
        opt.zero_grad()

    # 返回损失函数，样本个数（计算总损失时使用）
    return loss.item(), len(xb)
```

### 三行搞定！

```python
# 初始化模型化数据
train_dl, valid_dl = get_data(train_ds, valid_ds, bs)
# 得到模型，优化器
model, opt = get_model()
# 开始训练初始化损失函数
fit(20, model, loss_func, opt, train_dl, valid_dl)
```

```python
# 计算验证集数据的准确率
# 正确样本数
right_nums = 0
# 所有样本数
total_nums = 0
for xb, yb in valid_dl:
    # 拿到训练值
    outputs = model(xb)
    # 拿到预测值
    # 拿到训练值中可能性最大的种类的值和其所在位置
    _, predicted_index = torch.max(outputs.data, 1)
    total_nums += yb.size(0)
    # 拿到训练值与预测值一致的样本个数
    right_nums += (predicted_index == yb).sum().item()
    
print('准确率为：', right_nums / total_nums)
```

```python
# 测试数据集共有 79 x 128 + 16 = 10000 个测试数据集
# 遍历 79 个测试数据包
for xb, yb in valid_dl:
    # 拿到训练值
    outputs = model(xb)
    # 拿到预测值
    # 拿到当前训练包中 128 个训练数据中可能性最大的种类的值和其所在位置
    _, predicted_index = torch.max(outputs.data, 1)
    # 拿到训练结果与预测结果不符合的案例
    # 遍历每个数据包内的 128 个数据
    for index in range(predicted_index.nelement()):
        # 找到预测值与真实值不相等的测试数据
        if predicted_index[index] != yb[index]:
            pyplot.imshow(xb[index].reshape((28, 28)), cmap="gray")
            pyplot.show()
            # 预测值
            print("该测试数据的预测值为: ", predicted_index[index])
            print("该测试数据的实际值为: ", yb[index])
```
